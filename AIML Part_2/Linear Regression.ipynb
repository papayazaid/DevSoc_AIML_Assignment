```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.array([2, 3, 4, 5, 6, 7, 8, 9])
scores = np.array([45, 50, 55, 60, 65, 70, 78, 85])

x_mean = np.mean(hours)
y_mean = np.mean(scores)

numerator = np.sum((hours - x_mean) * (scores - y_mean))
denominator = np.sum((hours - x_mean)**2)
m = numerator / denominator
c = y_mean - m * x_mean

predicted_scores = m * hours + c

plt.scatter(hours, scores, label='Actual Scores')
plt.plot(hours, predicted_scores, color='red', label=f'Best Fit: y = {m:.2f}x + {c:.2f}')
plt.xlabel('Hours Spent on Event Preparation')
plt.ylabel('Coordinator Score')
plt.title('DevSoc Coordinator Performance')
plt.legend()
plt.show()

print(f"Slope (m): {m:.2f}")
print(f"Intercept (c): {c:.2f}")
```
